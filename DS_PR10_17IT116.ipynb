```
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```
img= cv2.imread('/ss.jpg')
plt.imshow(img[:,:,::-1])
plt.show()
```

analyze image

```
result= DeepFace.analyze(img,actions= ['gender'])
```

```
result_age= DeepFace.analyze(img,actions= ['age'])
```

results

```
plt.imshow(img[:,:,::-1])
print("gender:",result['gender'])
print("Age:",result_age['age'])
plt.show()
```

```
img2= cv2.imread('/jen.jpg')
plt.imshow(img2[:,:,::-1])
plt.show()
```

```
result= DeepFace.analyze(img2,actions= ['gender','age'])
```

```
plt.imshow(img2[:,:,::-1])
print("Gender:",result['gender'],"\nAge:",result['age'])
plt.show()
```
